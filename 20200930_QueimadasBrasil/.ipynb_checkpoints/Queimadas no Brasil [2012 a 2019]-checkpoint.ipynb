## Gráficos

Considerando os dados agrupados anteriormente, vamos mostrar um gráfico de barras, para representar a curva ABC.

Eu gosto bastante do gráfico de barras empilhadas, pois ele te dá uma ideia clara do que está acontecendo, sem firula.

Antes de plotar os gráficos, eu gosto de deixar umas variaveis carregadas, para que todos os gráficos tenham o mesmo tamanho.

## Analisando os maiores

Primeiramente vamos selecionar quantos (dos maiores) estados vamos analisar, e filtrar o dataset.

```python
n = 4
uf_big = queimadas_estados.sort_values(ascending=False)[:n].keys().values
dados_big = dados.query("UF in @uf_big")
```

veja que eu usei a "@" acima, para poder chamar o conteúdo de uma variavel e alimentar o filtro.

Como de prache, vamos olhar os 5 primeiros e os 5 últimos elementos

```python
uf_big
```

```python
dados_big
```

Vamos então agrupar a área queimada, por UF e ANO, para podermos plotar em um gráfico.

```python
queimadas_estados_anos = dados_big.groupby(by=["UF", "Ano"])['Area Queimada (mil)'].sum()
queimadas_estados_anos = queimadas_estados_anos.sort_values(ascending = False).reset_index()
queimadas_estados_anos.head()
```

Vamos plotar um gráfico de linhas, com cada estado em uma linha, para avaliar a evolução de cada um.

Como temos N estados, e queremos plotar uma linha para cada um, nós vamos usar um laço de repetição.
Veja bem: **uma linha para cada um**, logo, usaremos um "for" (para cada...) da lista "uf_big".

Já vimos que os dados tem que estar ordenados para ser plotados de forma correta, por isso, estamos usando o sort_values aqui também.

```python
titulo = "Evolução dos " + str(n) + " estados com maior área queimada"

plt.figure(figsize=((lado_min * fib), lado_min))

plt.title(titulo, weight='bold', size=tam_tit)
plt.ylabel("")
plt.xlabel("")


for i in uf_big:
    plt.plot(
        'Ano', 
        'Area Queimada (mil)', 
        data = queimadas_estados_anos.query("UF == @i ").sort_values(by = ['Ano']), 
        label = i,
        marker='o',
        linewidth=2
    )

plt.legend()

plt.show()
```

Olhando a evolução do gráfico de "MT" por exemplo, me faz entender que os casos de 2018 foram registrados ainda em 2017, pois o crescimento de 2017 frente 2016 foi expressivo, e depois caindo ao menor nivel histórico.

Algo que não podemos negligenciar é a analise da média móvel dos anos, aonde podemos querer ver a media móvel histórica, ou, móvel do ano em apigrafe + n anteriores.

Para iniciar a implementação, temos que calcular a média móvel, depois, basta plotar.

## Analisando a Média Móvel dos Maiores

Vamos então definir uma função para fazer tal cálculo, e depois, basta invocá-la no dataset contendo os dados a analisar.

Sademos que uma média móvel pega a soma acumulada e divide por um periodo estabelecido, aonde o efeito é uma suavização da curva, mostrando a média do ano em epigrafe + n anteriores.

Só que, se seus dados começaram em 2012 (como nosso caso aqui), e define-se que n = 5, logo, não teremos dados para 2007 a 2011, logo, mesmo sendo periodo = 5, temos que considerar nesse caso, 2012 a 2015 = 4 anos.

```python
def media_movel(uf, ano, n):
    
    tabela   = queimadas_estados_anos
    ano_max  = str(int(ano))
    ano_min  = str(int(ano) - n + 1)
    tabela   = tabela.query('UF == @uf and Ano <= @ano_max and Ano >= @ano_min ')
    anos     = len(tabela['Ano'].unique())
    valor    = tabela['Area Queimada (mil)'].sum()
    media    = valor / anos
    return     media

# testando função
media_movel('MG', 2014, 2)
```

Com a função definida, basta definir o periodo de média móvel, e criar a série (coluna) no dataframe (tabela).

```python
# PERIODOS MÉDIA MÓVEL
periodo_media_movel = 4
```

```python
queimadas_estados_anos['Area Queimada (mil) MM'] = queimadas_estados_anos.apply(lambda x: media_movel(x['UF'], x['Ano'], periodo_media_movel), axis=1)
queimadas_estados_anos.head(10)
```

Você percebeu que no exemplo acima usamos o "queimadas_estados_anos.apply" ou seja, nosso x do lamda é o dataset queimadas_estados_anos, logo, podemos usar as séries dele dentro do lambda e passar para a função.

Agora está fácil, basta usar o mesmo script do gráfico anterior, alterando a variável quantitativa, e assim teremos a curva que desejamos.

```python
titulo = "Evolução da Média Móvel de área queimada"

plt.figure(figsize=((lado_min * fib), lado_min))

plt.title(titulo, weight='bold', size=tam_tit)
plt.ylabel("")
plt.xlabel("")


for i in uf_big:
    plt.plot(
        'Ano', 
        'Area Queimada (mil) MM', 
        data = queimadas_estados_anos.query("UF == @i ").sort_values(by = ['Ano']), 
        label = i,
        marker='o',
        linewidth=2
    )

plt.legend()

plt.show()
```

Perceba a diferença, evidenciando que a situação no MT (estado do Mato Grosso) estava estabilizada (não controlada) entre 2013 a 2016, mas depois disso voltou a crescer.

Mas também, conseguimos ver que de 2017 a 2019 está havendo queda em PI (estado do Piauí), batendo em 2019 a menor série histórica, e possivelmente, devendo encontrar MT no gráfico em 2020.

## A representatividade frente ao tamanho da reserva

Queremos agora ver o quanto isso representa do total da área estabelecida como reserva.

Nesse caso, vou comparar a soma acumulada da série histórica, com o valor declarado como reserva.

Saliento aqui que meu foco não é ver se o dataset está certo, errado ou omisso, e sim, de apresentar dados com o conjunto de dados que nos foi entregue.

Para fazer esse cálculo, temos que pegar os valores unicos das áreas, e isso faremos da seguinte maneira:

```python
tamanho_reservas_UF = dados_big.groupby(by=["Código CNUC", "UF"])['Área estimada da UC (mil ha)'].max().reset_index()
tamanho_reservas_UF.head()
```

Agora basta agrupar os dados por estado...

```python
tamanho_reservas = tamanho_reservas_UF.groupby(by=["UF"])['Área estimada da UC (mil ha)'].sum()
tamanho_reservas.head()
```

Vamos fazer a soma acumulada agora de outra maneira, acumulando por UF e Ano.

```python
acumulado = dados_big.groupby(by=['UF', 'Ano'])['Area Queimada (mil)'].sum()
acumulado = acumulado.groupby(level=0).cumsum().reset_index()
acumulado = acumulado.set_index('UF').join(tamanho_reservas)
acumulado['percentual'] = acumulado['Area Queimada (mil)'] / acumulado['Área estimada da UC (mil ha)'] * 100

# mostrar somente dados de Minas Gerais
acumulado.query("UF == 'MT'")
```

```python
titulo = "Evolução do % de área queimada x tamanho da reserva"

plt.figure(figsize=((6 * 1.618), 6))

plt.title(titulo, weight='bold', size=14)
plt.ylabel("")
plt.xlabel("")


for i in uf_big:
    plt.plot(
        'Ano', 
        'percentual', 
        data = acumulado.query("UF == @i ").sort_values(by = ['Ano']), 
        label = i,
        marker='o',
        linewidth=2
    )

plt.legend()

plt.show()
```

## Tentando identificar aonde pode estar o problema (nos estados analisados)

```python
reserva_uf_tamanho = dados_big.groupby(by=["Código CNUC", "UF"])['Área estimada da UC (mil ha)'].max()
reserva_uf_queimada = dados_big.groupby(by=["Código CNUC", "UF"])['Area Queimada (mil)'].sum()
reserva_uf_consolidada = pd.DataFrame(reserva_uf_tamanho).join(reserva_uf_queimada)
reserva_uf_consolidada['percentual'] = reserva_uf_consolidada['Area Queimada (mil)'] / reserva_uf_consolidada['Área estimada da UC (mil ha)'] * 100
```

Olha só como é facil, denovo, fazer o "procv" entre duas séries (colunas) do Pandas.

Basta que elas tenham o mesmo indice, e basta aplicar o JOIN

```python
reserva_uf_consolidada = reserva_uf_consolidada.sort_values(by = ['percentual'], ascending = False)

# filtrando as reservas com queimadas acima de 100%
reserva_uf_consolidada.query('percentual > 100')
```

Conseguimos identificar, e evidenciar, aqui algumas reservas com percentual queimado maior que a metragem total dela.

Mas por que?

Não temos mais informação sobre como esses dados são informados, para ser sumarizados, mas, algo estranho há, pois parece que estão usando algumas reservas como espécie de "curva de rio", ou seja, não sei aonde lançar, lanço lá!

## E quanto sobra de área verde nos registros?

```python
# Total da área queimada (em mil hectares), apresentado no dataset
area_total_queimada = dados['Area Queimada (mil)'].sum()
area_total_queimada
```

```python
# Total da área de reserva (em mil hectares), apresentado no dataset
area_total = tamanho_reservas.sum()
area_total
```

```python
# Percentual queimado
per_queimado = round(area_total_queimada / area_total * 100,0) 
```

Vamos usar o gráfico de Waffle para mostrar qual a área queimada, da total, e para isso, vamos criar um dicionário com os valores a plotar, e depois é só plotar o gráfico.

```python
data = {
    "Reserva Virgem" : (area_total - area_total_queimada),
    "Area Queimada" : area_total_queimada
}
data
```

```python
# plotar gráfico de waffle
fig = plt.figure(
                FigureClass=Waffle,
                rows=5,
                columns=15,
                colors=("#90EE90", "#FF0000"),
                values=data, 
                icons='tree', 
                icon_size=30,
                legend={'loc': 'upper left', 'bbox_to_anchor': (1,1)},
                icon_legend=True,
                figsize=((6 * 1.618), 6)
                )
fig.set_tight_layout(False)
titulo = "De 2012 a 2019 foi queimado " + str(per_queimado) + "% Do total da reserva declarada"
plt.title(titulo, weight='bold', size=14)
plt.show();
```

```python
reserva_uf_tamanho = dados.groupby(by=["Bioma referencial"])['Área estimada da UC (mil ha)'].max()
reserva_uf_queimada = dados.groupby(by=["Bioma referencial"])['Area Queimada (mil)'].sum()
reserva_uf_consolidada = pd.DataFrame(reserva_uf_tamanho).join(reserva_uf_queimada)
reserva_uf_consolidada['percentual'] = reserva_uf_consolidada['Area Queimada (mil)'] / reserva_uf_consolidada['Área estimada da UC (mil ha)'] * 100
reserva_uf_consolidada.sort_values(by = ['percentual'], ascending = False)
```

Fica aqui, mais uma vez evidente, que estão sendo apontadas queimadas em área maior que as reservas naturais constituidas.
